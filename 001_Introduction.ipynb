```python
%config Completer.use_jedi = False
%autosave 20

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras

import plotly.express as px
import plotly.graph_objects as go
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train_flat=x_train.reshape(len(x_train),28*28)
x_test_flat=x_test.reshape(len(x_test),28*28)
```

```python
model=keras.Sequential([
        keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train_flat,y_train,epochs=5)
```

```python
model.evaluate(x_test_flat,y_test)
```

```python
y_pred=model.predict(x_test_flat)
y_pred
```

```python
px.imshow(x_test[12])
```

```python
y_pred[12]
```

```python
np.argmax(y_pred[12])
```

```python
conMat=tf.math.confusion_matrix(labels=y_test,predictions=[np.argmax(i) for i in y_pred])
conMat
```

```python
px.imshow(
    conMat,
    labels=dict(x="Predicted", y="Truth"),
    text_auto=True
)
```

```python
hlModel=keras.Sequential([
        keras.layers.Dense(100, input_shape=(784,), activation='relu'),
        keras.layers.Dense(10, activation='sigmoid')
])

hlModel.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

hlModel.fit(x_train_flat,y_train,epochs=5)
```

```python
hlModel.evaluate(x_test_flat,y_test)
```

```python
hlY_pred=hlModel.predict(x_test_flat)
```

```python
mlConMat=tf.math.confusion_matrix(labels=y_test,predictions=[np.argmax(i) for i in hlY_pred])
mlConMat
```

```python
px.imshow(
    mlConMat,
    labels=dict(x="Predicted", y="Truth"),
    text_auto=True
)
```

```python
hlModel1=keras.Sequential([
        keras.layers.Flatten(input_shape=(28,28)),
        keras.layers.Dense(120, activation='relu'),
        keras.layers.Dense(40, activation='relu'),
        keras.layers.Dense(10, activation='sigmoid')
])

hlModel1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

hlModel1.fit(x_train,y_train,epochs=10)
```
